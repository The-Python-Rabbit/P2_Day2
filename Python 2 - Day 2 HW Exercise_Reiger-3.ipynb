## Exercise 1: 
Create a Temperature Converter program using OOP by creating a Temperature class with two private attributes to store Fahrenheit and Celsius degrees. In the Temperature class, define methods that 
- sets the private attributes. When you set one unit of temperature, it should calculate and set the other unit of temperature. For example, when you set degrees in Fahrenheit, it should calculate and set in Celsius degrees. 
- gets the hidden attributes that round the number to 2 decimal places. 

The output should look something like following:

    MENU
    1. Fahrenheit to Celsius
    2. Celsius to Fahrenheit
    3. Quit
    
    Enter a menu option: 1
    Enter degrees in Fahrenheit: 99
    99.00 oF is 37.22 oC.
    
    Enter a menu option: 2
    Enter degrees in Celsius: 37.22
    37.22 oC is 99.00 oF.
    
    Enter a menu option: 3
    Bye


```python
class Temperature:
    def __init__(self):
        self.__temp = 0.00
    
    def f_to_c(self,temp):
        self.__temp = temp
        return f"{self.__temp} oF is {((self.__temp-32)/1.8):,.2f} oC"
    
    def c_to_f(self,temp):
        self.__temp = temp
        return f"{self.__temp} oC is {(1.8*self.__temp+32):,.2f} oF"
    
temp = Temperature()
print ("MENU")
print ("\n1. Fahrenheit to Celsius")
print ("\n2. Celsius to Fahrenheit")
print ("\n3. Quit")

choice = 1
 
while choice:    
    choice = int(input("\nEnter a menu option: "))
    if choice == 1:
        t=float(input("Enter degrees in Fahrenheit: "))
        print (temp.f_to_c(t))
    
    elif choice == 2:
        t=float(input("Enter degrees in Celsius: "))
        print (temp.c_to_f(t))
    
    elif choice == 3:
        break
    
    else:
        print ("Please enter 1, 2 or 3")
    
print ("\nGoodbye. Have a cool day!")
        
```

```python
# From Day 1 Class Demo
class Product : 
    """A simple attempt to model a product."""
    
    def __init__(self, name, price, discount_rate, description) :
        """Initialize name, price, and discount_rate attributes"""
        self.name = name
        self.price = price
        self.discount_rate = discount_rate
        self.description = description
        
        
    def get_discount_amount(self) :
        """Computes a discount calculation"""
        return self.price * self.discount_rate / 100
    
        
    def get_info(self) :
        return(f"Name: {self.name}\n" + 
                f"Price: {self.price}\n" +
                f"Disicount Amount: ${self.get_discount_amount():,.2f}\n" +
                f"Discounted Price: ${self.get_sale_price():,.2f}")
    
    def get_sale_price(self) :
        """Calls another method to find a sale price"""
        return self.price - self.get_discount_amount()
    
    def get_description(self):
        """Add get_description method for this exercise"""
        return (f"{self.name} is a:\n" +
                 f"{self.description}\n")
    
class Book(Product):
    """Create the Book class, which is a subset of the Product class. But don't use all of attributes from Product class"""
    def __init__(self, name, price, discount_rate, description, author):
        super().__init__(name, price, discount_rate, description)
        self.author = author
    
    """Get description info associated with the Product class and then 'author' information unique to Book class"""
        
    def get_a_description(self):
        return (f"\n{super().get_description()}by {self.author}")
    
class Movie(Product):
    """Create the Movie class, which is a subset of the Product class. But don't use all of the attributes."""
    
    def __init__(self, name, price, discount_rate, description, year):
        super().__init__(name, price, discount_rate, description)
        self.year = year
        
    """Get description info associated with the Product class and then 'year' information unique to Movie class"""        
        
    def get_m_description(self):
        return (f"\n{super().get_description()}that was released in {self.year}")
        
book1 = Book("The Grapes of Wrath", 39.95, 90, "fiction book", "John Steinbeck")
print (book1.get_a_description())

movie1 = Movie("2001: A Space Odyessy", 0,0, "science-fiction movie", "1968")
print (movie1.get_m_description())
        
```

```python
class MyDate:
    
    def __init__(self,date_string):
        
        self.list = []
        self.set_datetime(date_string)
        
# ------------------------------------------------------------------------------------------------
      
    def get_datetime(self, month_first = True):
        """Get the date and time but print error message if invalid data was entered"""

        self.__month_first = month_first
        months = {1:"January",2:"February", 3:"March", 4:"April", 5:"May", 6:"June",
                  7:"July", 8:"August", 9:"September", 10: "October", 11:"November", 12:"December"}
        
# Function "set_datetime" assigns -1 to invalid data. So print error message if data is invalid

        if self.__month == -1 or self.__day == -1 or self.__hour == -1 or self.__min == -1 or self.__sec == -1:
            return ("\nCannot get date and time because invalid data was entered.\n")
        else:
            return (f"\nThe date is {months[self.__month]} {self.__day}, {self.__year}." +
                f"\nThe time is {self.__hour} hours, {self.__min} minutes, {self.__sec} seconds.")
        
# -----------------------------------------------------------------------------------------------------
        
    def set_datetime(self, date_string):
        """Process the input string into a list and assign data to variables"""
        
        date_string = date_string.replace("-"," ")
        date_string = date_string.replace(":", " ")

        self.list = date_string.split()

# Check data. Assign data to variables after validation.
# Assign place holder value of -1 if user entered invalid data


        if int(self.list[0]) < 1 or int(self.list[0]) > 12:
            print (f"\nMonth must be 1 to 12. {int(self.list[0])} is invalid")
            self.__month = -1
        else:
            self.__month = int(self.list[0])

# Check if day is valid for month

        if self.__month != -1:
        
# Check for February. Do not check for leap year.

            if self.__month == 2:
                if int(self.list[1]) < 1 or int(self.list[1]) > 28:
                    print (f"\nDay must be 1 to 28. {int(self.list[1])} is invalid")
                    self.__day = -1
                else:
                    self.__day = int(self.list[1])
                
# Check for April, June, September or November

            elif self.__month == 4 or self.__month == 6 or self.__month == 9 or self.__month == 11:
                if int(self.list[1]) < 1 or int(self.list[1]) > 30:
                    print (f"\nDay must be 1 to 30. {int(self.list[1])} is invalid")
                    self.__day = -1
                else:
                    self.__day = int(self.list[1])
                
# Check for January, March, May, July, August, October or December               
                
            elif self.__month == 1 or self.__month == 3 or self.__month == 5 or self.__month == 7 or self.__month == 8 or self.__month == 10 or self.__month == 12:
                if int(self.list[1]) < 1 or int(self.list[1]) > 31:
                    print (f"\nDay must be 1 to 31. {int(self.list[1])} is invalid")
                    self.__day = -1
                else:
                    self.__day = int(self.list[1])

        self.__year = int(self.list[2])
        
# Check for valid hour
        
        if int(self.list[3]) < 1 or int(self.list[3]) > 24:
            print (f"\nHour must be 1 to 24. {int(self.list[3])} is invalid")
            self.__hour = -1
        else:
            self.__hour = int(self.list[3])

# Check for valid minute
 
        if int(self.list[4]) < 1 or int(self.list[4]) > 60:
            print (f"\nMinute must be 1 to 60. {int(self.list[4])} is invalid")
            self.__min = -1
        else:
            self.__min = int(self.list[4])
            
# Check for valid second
            
        if int(self.list[5]) < 1 or int(self.list[5]) > 60:
            print (f"\nSeconds must be 1 to 60. {int(self.list[5])} is invalid")
            self.__sec = -1
        else:
            self.__sec = int(self.list[5])
            
# --------------------------------------------------------------------------------------
    def add_hour(self, new_hour):
        
        """Function to add hours to the existing hour. Does not account for leap years """
   
        new_total_hours = self.__hour + new_hour

        """add years if necessary"""
    
        if new_total_hours > 24:
            add_days = new_total_hours//24
            self.add_days(add_days)
        
        self.__hour = new_total_hours % 24
  
 # --------------------------------------------------------------------------------------
    def add_days (self, days):
        """Function to add days to current day. Does not account for leap year"""
    
        """Store total days for the year up to the end of the previous month"""
        current_total_days = {1:0,2:31, 3:59, 4:90, 5:120, 6:151, 7:181, 8:212, 9:243, 10:273, 11:304, 12:334}
        
        new_total_days = current_total_days[self.__month] + self.__day + int(days)

        """Determine if any years must be added to current year"""
        
        if new_total_days > 365:
            add_years = new_total_days//365
            self.__year += add_years

        """Calculate the day of the year for the new date"""    
            
        new_total_days %= 365
        
        if new_total_days > 334:
            self.__month = 12
            self.__day = new_total_days - 334
        elif new_total_days > 304:
            self.__month = 11
            self.__day = new_total_days - 304  
        elif new_total_days > 273:
            self.__month = 10
            self.__day = new_total_days - 273  
        elif new_total_days > 243:
            self.__month = 9
            self.__day = new_total_days - 243  
        elif new_total_days > 212:
            self.__month = 8
            self.__day = new_total_days - 212  
        elif new_total_days > 181:
            self.__month = 7
            self.__day = new_total_days - 181  
        elif new_total_days > 151:
            self.__month = 6
            self.__day = new_total_days - 151  
        elif new_total_days > 120:
            self.__month = 5
            self.__day = new_total_days - 120  
        elif new_total_days > 90:
            self.__month = 4
            self.__day = new_total_days - 90  
        elif new_total_days > 59:
            self.__month = 3
            self.__day = new_total_days - 59  
        elif new_total_days > 31:
            self.__month = 2
            self.__day = new_total_days - 31  
            
        else:
            self.__month = 1
            self.__day = new_total_days 
            
#   *************************************************************************************************************

today = MyDate("06-21-2021 17:51:33")
          
print (today.get_datetime())
print (today.get_datetime(True))
print (today.get_datetime(month_first=False))
today.set_datetime("11-31-1999 02:33:22")

# Enter invalid values to verify data validation

# Invalid month
today.set_datetime("13-31-1999 02:33:22")
print (today.get_datetime())

# Invalid day

today.set_datetime("7-46-1999 02:33:22")
print (today.get_datetime())

# Invalid hour
today.set_datetime("11-31-1999 99:33:22")
print (today.get_datetime())

# Invalid minute
today.set_datetime("11-31-1999 02:66:22")
print (today.get_datetime())

# Invalid second
today.set_datetime("11-31-1999 02:33:67")
print (today.get_datetime())
#today.set_datetime("06-22-2021 13:36:03")
#print (today.get_datetime())

#today.add_hour(14)
#print (today.get_datetime())

#today.add_days(10)
#print (today.get_datetime())
#today.set_datetime("06-22-2021 13:36:03")
#today.add_days(50)
#print (today.get_datetime())
#today.set_datetime("06-22-2021 13:36:03")
#today.add_days(150)
#print (today.get_datetime())

today.set_datetime("06-22-2021 13:36:03")
today.add_hour(4)
print (today.get_datetime())

today.set_datetime("06-22-2021 13:36:03")
today.add_hour(34)
print (today.get_datetime())

today.set_datetime("06-22-2021 13:36:03")
today.add_hour(340)
print (today.get_datetime())

today.set_datetime("06-22-2021 13:36:03")
today.add_hour(34000)
print (today.get_datetime())

today.set_datetime("06-22-2021 13:36:03")
today.add_days(1000)
print (today.get_datetime())
```
